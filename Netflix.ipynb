```python
movies_df.drop(columns=['date_added','release_year', 'rating', 'duration'],inplace=True)
```

```python
movies_df.dropna(inplace=True)
```

```python
movies_df["show_id"]=movies_df["show_id"].str.replace("s","")
```

```python
movies_df.reset_index(drop=True, inplace=True)
```

```python
movies_df
```

```python
#import pickle
#pickle.dump(movies_df,open("clean_movie.pkl","wb"))
```

```python
movies_df["type"] = movies_df["type"].apply(lambda x: x.replace(" ", "").lower())
movies_df["cast"] = movies_df["cast"].apply(lambda x: x.replace(" ", "").lower())
movies_df["country"] = movies_df["country"].apply(lambda x: x.replace(" ", "").lower())
movies_df["listed_in"] = movies_df["listed_in"].apply(lambda x: x.replace(" ", "").lower())
movies_df["director"] = movies_df["director"].apply(lambda x: x.replace(" ", "").lower())
movies_df["description"] = movies_df["description"].str.lower()
```

```python
movies_df
```

```python
movies_df["tags"]=movies_df["type"]+","+ movies_df["director"]+","+movies_df["cast"]+","+movies_df["country"]+","+movies_df["listed_in"]+","+movies_df["description"]
```

```python
movies_df
```

```python
movies_df.drop(columns=["type","director","cast","country","listed_in","description"],inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps=PorterStemmer()
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer  # term frequency- inverse document frequency
```

```python
movies_df["tags"]=[lemma.lemmatize(word) for word in movies_df["tags"] if word not in set(stopwords.words("english"))]
```

```python
movies_df
```

```python
# vectorizer
vectorizer=TfidfVectorizer()
feature_vectors=vectorizer.fit_transform(movies_df["tags"])
print(feature_vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(feature_vectors)
print(similarity)
```

```python
movie_list= movies_df["title"].tolist()
movie_list
```

```python
def recommend(movie):
    movie_list= movies_df["title"].tolist()
        # find the close match for the movie name given by the user
    import difflib
    find_close_match=difflib.get_close_matches(movie,movie_list)
    print(find_close_match)
    close_match=find_close_match[0]
    # finding the indx of the movie with title
    movie_index= movies_df[movies_df.title==close_match].index.values[0] 
    # sorting the movie based ob their similarity score
    sort_similar_movies=sorted(list(enumerate(similarity[movie_index])),key= lambda x:x[1],reverse=True)
    # Print the name of the similar movie based on tthe index
    print("Movies sugeested for you: \n")
    i=1
    for movies in sort_similar_movies:
        index=movies[0]
        title_from_index = movies_df[movies_df.index==index]["title"].values[0]
        if(i<=20):
            print(i,".",title_from_index)
        i+=1
    return 
```

```python
def get_show_details_by_title(df, title):
    
    result = df[df['title'] == title]
    
    if result.empty:
        return []  # Return an empty list if no match is found
    else:
        details = []
        for _, row in result.iterrows():
            details.append({
                "Show ID": row['show_id'],
                "Type":row["type"],
                "Description": row['description'],
                "Cast": row['cast'],
                "Genre": row['listed_in']
            })
        return details
```
